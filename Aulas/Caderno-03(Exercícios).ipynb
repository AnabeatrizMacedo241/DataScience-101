```python
#1
import pandas as pd
import matplotlib.pyplot as plt
euro = pd.read_csv('euro2020.csv', sep=',')
euro.fillna(method='bfill', inplace=True) #Remover NaN
euro.set_index('Player', inplace=True)
euro
```

```python
euro.info()
```

```python
euro.describe()
```

```python
#2
euro['Goals'].plot(kind='density')
```

```python
#3
euro['Position'].value_counts() #M = Middle, F = Foward, D = Defense
```

```python
euro['Position'].value_counts().plot(kind='pie')
```

```python
#4
euro['Goals_per_attempt'] = euro['Goals'] / euro['Total attempts']
euro['Goals_per_attempt'].plot(kind='density', figsize=(15,7))
```

```python
#Outro jeito, mas sem ter que fazer a conta
euro.plot(kind='scatter', x='Goals', y='Total attempts', title='Gols por tentativas')
```

```python
#5
stats = ['Goals', 'Right foot goals', 'Left foot goals', 'Total attempts', 'On target', 'Passing accuracy']

euro[stats].head(10).plot(kind='bar', subplots=True, layout=(2,3), figsize=(20,8))
```
