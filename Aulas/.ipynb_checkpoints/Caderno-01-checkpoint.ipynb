## O que é Data Science?
Data Science, conhecida no Brasil como Ciência de Dados, é o processo de análise de dados que nos ajudará a tomar decisões mais precisas a fim de obter melhores resultados em diversos campos como, por exemplo: logística, política, indústria, ciência etc...

Para aprendermos Data Science, precisamos entender um pouco de programação. As duas linguagens mais usadas para a análise de dados é o Python e R. Em meu repositório há aulas e tutoriais sobre Python caso tenham interesse em aprender mais. Link:https://github.com/AnabeatrizMacedo241/Python-101

```python
import pandas as pd #Usamos o pandas(biblioteca) para auxiliar na extração, tratamento e análise dos dados.
tabela = {'coluna1': ['Ana', 'João', 'Maria', 'Pedro', 'Isabela'], 'coluna2': [4, 5, 6, 9, 5], 'coluna3': [7, 8, 12, 1, 11]}
df = pd.DataFrame(data=tabela)
print(df)
```

Documentação do Pandas: https://pandas.pydata.org/docs/reference/frame.html . É sempre bom pesquisar na internet ou na documentação quando surge um problema ou dúvida.

```python
num_colunas = df.shape[1] #Podemos contar o número de colunas. Colunas são representadas pelo index 1
print(num_colunas)
```

```python
num_linhas = df.shape[0] #As linhas são representadas pelo index 0
print(num_linhas)
```

```python
df.max() #max() retorna os valores mais altos de cada coluna
```

```python
df.min() #min() retorna os menores valores de cada coluna
```

```python
df.mean() #mean() é a média dos valores de cada coluna
```

```python
df.sum() #Aqui será somado todos os valores
```

### E se quisermos analisar uma tabela já exsitente? 
Irei usar esse dataset como exemplo para mostrar como importá-lo e depois analisar seus dados. Vocês consegue encontrar esse site através desse linnk:https://www.kaggle.com/preetviradiya/imdb-movies-ratings-details.

```python
import pandas as pd
imdb_review = pd.read_csv('IMDB_movie_reviews_details.csv', header=0, sep=',')
imdb_review
```

1. Nomeamos os dados
2. pd.read_csv serve para abrir um arquivo csv
3. header=0 significa que a primeira linha da tabela será o nome das nossas colunas.
4. O sep=','significa que cada informação será divdidida pela vírgula. Caso algum não esteja separado, teremos que separar, porque se não o código não funcionará.

```python
imdb_review.describe() #Aqui podemos 'resumir' os dados
```

Podemos observar que há informações estranhas como `NaN` que significa: 'not a number'(não é um número). Como resolver esse problema?

```python
imdb_review.dropna(axis=0,inplace=True)
print(imdb_review) #Usamos dropna para remover os 'NaN' de todas as linhas(axis=0)
```

```python
imdb_review #Veja que os 'NaN' não estão mais na tabela
```

```python
imdb_review.info() #Com info() podemos descobrir todos os nomes das colunas e seu tipo de valor(int, float, object, boolean)
```

```python
#Caso algum Dtype não esteja correto, podemos convertê-lo
imdb_review['votes'] = [x.replace(',', '.') for x in imdb_review['votes']]
imdb_review['gross'] = [x.replace('M', '') for x in imdb_review['gross']]
imdb_review['gross'] = [x.replace('$', '') for x in imdb_review['gross']
#Mudamos a vírgula da coluna votes para um ponto, porque a vírgula não permite fazer operações.
#Mudamos o 'M' de milhão para nada('') e '$', porque ele também não permite realizar operações por ser string.
```

```python
imdb_review
```

```python
del imdb_review['Unnamed: 0']
#Aqui eu consegui deletar a coluna que não estava nos ajudando
```

```python
imdb_review
```

```python
imdb_review.head() #head() retorna as primeiras linhas da tabela
```

```python
imdb_review[0:3] #Aqui podemos selecionar linhas específica com o slicing
```

```python
select = imdb_review.genre == 'Drama'
imdb_review[select]
#Aqui eu filtro todos os filmes que são do genêro drama
```

```python
imdb_review[select].mean() #Aqui retornaram as médias de cada coluna dos filmes dramáticos
```

```python
imdb_review[(imdb_review.genre == 'Drama') & (imdb_review.year == 2001)]
```

`loc.` Seleciona um grupo de linhas e colunas segundo os rótulos ou uma matriz booleana. 
`iloc.` Seleciona com base nos índices, ou seja, se baseia na posição das informações.

```python
imdb_review.loc[55]
```

```python
imdb_review.iloc[[-1]] #Aqui ele retornou a última linha da tabela
```

```python
imdb_review.iloc[-1, [1, 5, 2]]
#Aqui da última linha, retornamos as colunas 1(name), 5(rating) e 2(year)
```
